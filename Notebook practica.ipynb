```python
import networkx as nx
import matplotlib.pyplot as plt

```

```python
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_edge(1,2)
G.add_nodes_from([3,4,5])
G.add_edges_from([(1,4),(3,4),(2,4),(3,2),(3,5)])
print(G)
print(G.nodes)
print(G.edges)
print(list(G.neighbors(2)))
```

```python
G.remove_node(5)
print(G.nodes)
print(G.edges)
```

```python
G[1]# Adyacencia de 1
G[2]# Adyacencia de 2
G[1][2]# Contenido de la arista 12
G.adj
```

```python
plot = plt.plot()
nx.draw(G,with_labels=True,pos={1:(0,0),2:(0,1),3:(1,0),4:(1,1)})
plt.show()
```

```python
G2=nx.Graph()
G2.add_nodes_from(['a','b',10,11,3.14])
G2.add_edge('a',10,object="Una cadena",weight=1.1)
G2.add_edges_from([('a',10),('a',3.14),('b',11)])
print(G2)
print(G2['a'][10])
print(G2['a'][10]['weight'])
plot=plt.plot()
nx.draw(G2,with_labels=True)
plt.show()
```

```python
Gd = nx.DiGraph()
Gd.add_nodes_from([1,2,3,4])
Gd.add_edges_from([(1,2),(2,1),(1,4),(3,4),(2,4),(4,2),(3,2)])
plot=plt.plot()
nx.draw(Gd,with_labels=True)
plt.show()
```

```python
print(G.degree())
print(Gd.degree())
print(Gd.in_degree())
print(Gd.out_degree())
print(Gd.in_degree(1),Gd.out_degree(1))
```

```python
componentes = nx.connected_components(G)
print(list(componentes))
componentes = nx.connected_components(G2)
print(list(componentes))
componentes_debil = nx.weakly_connected_components(Gd)
print(list(componentes_debil))
componentes_fuerte = nx.strongly_connected_components(Gd)
print(list(componentes_fuerte))
```

```python
aam = nx.minimum_spanning_tree(G)
print(aam.nodes)
print(aam.edges)
```

```python
plot = plt.plot()
pos = nx.spring_layout(G)
nx.draw(G,pos=pos,with_labels = True,edge_color = 'b')
nx.draw(aam,pos=pos,with_labels = True,edge_color = 'r')
plt.show()
```

```python
camino = nx.shortest_path(G,1,3)
print(camino)

```

```python
plot = plt.plot()
pos = nx.spring_layout(G)
nx.draw(G,pos=pos,with_labels = True,edge_color = 'b')
nx.draw_networkx_edges(G,pos=pos,arrows = True,edgelist = [(1,4),(4,3)],edge_color = 'r')
nx.draw_networkx_edge_labels(G,pos=pos,edge_labels={(1,4):'1',(4,3):'2'})
plt.show()
```
